#' Calculate the factorial of a number
#'
#' This function computes the factorial of a given non-negative integer `n`.
#' The factorial of `n` is defined as the product of all positive integers
#' less than or equal to `n`. By definition, the factorial of 0 is 1.
#'
#' @param n A non-negative integer for which the factorial is to be calculated.
#' @return The factorial of the input number `n`.
#' @examples
#' calculate_factorial(5) # Returns 120
#' calculate_factorial(0) # Returns 1

    if (n == 0) {
        return(1)
    } else {
        return(n * calculate_factorial(n - 1))
    }
